```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
base_url = "https://uk.trustpilot.com/review/myurbanjungle.com?verified=true"

page = requests.get(base_url)
soup = BeautifulSoup(page.text, 'html')
```

```python
reviews = soup.find_all('p', class_ = 'typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn')
cleaned_reviews = [review.get_text(strip=True) for review in reviews]
cleaned_reviews
```

```python
stars = soup.find_all('img',attrs={'alt': re.compile(r'Rated [1-5] out of 5 stars')})
star_rating = []
for img in stars:
    alt_text = img['alt']
    # Split the string and extract the number
    rating = int(alt_text.split()[1])
    star_rating.append(rating)
star_rating
```

```python
review_dict = {}
for review,star in zip(cleaned_reviews,star_rating):
    review_dict[review] = star
  
review_dict
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
#Convert Dictionary into a list of Dictionaries
data = [{"Review": review, "Star Rating": star} for review, star in review_dict.items()]

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
page_number = 2
cleaned_reviews2 = []
while page_number <= 100:
    # Construct the URL for the current page
    url = "https://uk.trustpilot.com/review/myurbanjungle.com?page={}&verified=true".format(page_number)
    page2 = requests.get(url)
    soup2 = BeautifulSoup(page2.text, 'html')
    
    # Find the reviews on the current page
    reviews2 = soup2.find_all('p', class_ = 'typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn')
    
    for review in reviews2:
        cleaned_reviews2.append(review.get_text(strip=True))
    
    # Increment the page number for the next loop iteration
    page_number += 1
cleaned_reviews2
```

```python
page_number_2 = 2
star_rating_2 = []
while page_number_2 <= 100:
    # Construct the URL for the current page
    url_2 = "https://uk.trustpilot.com/review/myurbanjungle.com?page={}&verified=true".format(page_number_2)
    page_2 = requests.get(url_2)
    soup_2 = BeautifulSoup(page_2.text, 'html')
    
    # Find the rating on the current page
    stars_2 = soup_2.find_all('img',attrs={'alt': re.compile(r'Rated [1-5] out of 5 stars')})
    
    for img in stars_2:
      alt_text_2 = img['alt']
    # Split the string and extract the number
      rating_2 = int(alt_text_2.split()[1])
      star_rating_2.append(rating_2)
    
    # Increment the page number for the next loop iteration
    page_number_2 += 1
star_rating_2
```

```python
#Create Dictionary with reviews and ratings
review_dict2 = {}
for review,star in zip(cleaned_reviews2,star_rating_2):
    review_dict2[review] = star
  
review_dict2
```

```python
#Convert Dictionary into a list of Dictionaries
data2 = [{"Review": review, "Star Rating": star} for review, star in review_dict2.items()]

# Create DataFrame
df2 = pd.DataFrame(data2)
df2
```

```python
#Join DataFrames
df3=pd.concat([df, df2])
```
